# Abstract
Import, plot, and sort data from given coefficients of pressure around a NACA 4412 at Mach 0.3. Also perform numerical integration using the trapezoidal method.

## Import Data
Importing section

```python
import numpy as np
d = np.loadtxt("4412_Ma03Cp.txt",delimiter = "\t",skiprows = 1) # NumPy import method
x = d[:,0] # x-coordinates of aerfoil
z = d[:,2] # z-coordinates of aerfoil
cp = d[:,3] # Cp
# cp = np.flipud(cp) if you needed to flip vector
```

## Plot
Plotting section using 'matplotlib.pyplot'

```python
import matplotlib.pyplot as plt

# Plot:
fig,(ax1, ax2) = plt.subplots(1,2) # Create figure with subplots
ax1.plot(x,z,'k-',lw = 0.25) # Aerfoil plot
ax1.grid(True) 
ax1.set_title('Aerofoil Shape')
ax2.plot(x,cp,'b-') # Coefficient of pressure plot
ax2.grid(True)
ax2.set_title('Coefficient of Pressure vs X-location')
plt.gca().invert_yaxis() # Invert x-y axis
```

## Numerical Integration 
Discrete numerical integration using the trapezoidal method

```python
from trapz import trapz # Import made trapezoidal integration method

a = trapz(cp,x) # Perform integration
print(a) # Display value
```

## Data Sorting
Sort data here
Example used is for cp > 0.5, store values in a new array/vector

```python
greater = cp > 0.5 # Check cp values over 0.5
S = np.sum(greater) # Total that amount

j = -1 # Initialize j
greater = np.linspace(1,S,S) # Initialize 'greater'
for i in range(len(cp)): # Iterate over entire 'cp'
    if cp[i] > 0.5: # If cp(i) > 0.5, store that value as greater(j)
        j = j + 1 # Update 'j'
        greater[j] = cp[i] # Store value into 'greater'

print(greater) # Display 'greater'
```
