```python
import numpy as np
import matplotlib.pyplot as plt
# %matplotlib inline # No idea what this does
```

```python
a = np.linspace(0,2*np.pi,100) # Start,stop,size
b = np.sin(4*a)*np.exp(-a) # sin(4x)exp(-x)
c = np.cos(4*a)*np.exp(-a) # cos(4x)exp(-x)
```

```python
## Plot
plt.figure(1)
plt.plot(a,b,'bo-',label = '$e^{-x}sin(4x)$',linewidth = 1)
plt.plot(a,c,'ro-',label = '$e^{-x}cos(4x)$',linewidth = 1)
plt.grid(color = 'b',ls = '-.', lw = 0.25) # plt.grid(True) is the basic one
plt.xlabel('x')
plt.ylabel('y')
# plt.legend(['b','c']) # If you don't define the label in the plot line
plt.title('Title')
plt.legend()
plt.savefig('Figure 1.png') # Save image as .png
```
